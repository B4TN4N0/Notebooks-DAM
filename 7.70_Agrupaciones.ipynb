# üìì Cap√≠tulo 23: `groupby()`## üí° Introducci√≥nEl m√≥dulo **`itertools`** de Python es una colecci√≥n de herramientas para trabajar con iteradores de manera eficiente. La funci√≥n **`itertools.groupby(iterable, key=None)`** es particularmente √∫til para agrupar elementos adyacentes que comparten la misma 'clave' o 'criterio'.‚ö†Ô∏è **Advertencia Importante:** `groupby()` solo agrupa elementos *adyacentes*. Si los elementos con la misma clave no est√°n contiguos, se crear√°n grupos separados. Por lo tanto, **siempre se debe ordenar el iterable primero** si se desea agrupar todos los elementos con la misma clave.

## 23.1. Ejemplo B√°sico 1: Agrupaci√≥n SimpleEn este ejemplo, definiremos una funci√≥n clave simple para agrupar n√∫meros por su paridad (par o impar).

```python
from itertools import groupbydata_sin_orden = [1, 2, 2, 3, 3, 3, 4, 4, 5]# Definir la funci√≥n clave (el criterio de agrupaci√≥n)# En este caso, el valor del elemento es la claveprint("--- Agrupaci√≥n por valor --- \n")for key, group in groupby(data_sin_orden):    # group es un iterador; usamos list() para ver todos los elementos    print(f"Clave: {key}, Elementos: {list(group)}")
```

## 23.3. Ejemplo 3: Agrupaci√≥n de Diccionarios o TuplasUsaremos una funci√≥n clave para agrupar una lista de tuplas o diccionarios bas√°ndonos en un campo o √≠ndice espec√≠fico.

```python
from itertools import groupbyimport operatorpersonas = [    {'nombre': 'Alice', 'edad': 30, 'ciudad': 'NY'},    {'nombre': 'Bob', 'edad': 25, 'ciudad': 'LA'},    {'nombre': 'Charlie', 'edad': 30, 'ciudad': 'NY'},    {'nombre': 'Dave', 'edad': 25, 'ciudad': 'LA'},]# 1. Definir la funci√≥n clave: agrupar por 'edad'key_edad = operator.itemgetter('edad')# 2. Ordenar la lista por la clave de agrupaci√≥n ('edad')personas_ordenadas = sorted(personas, key=key_edad)print("--- Agrupaci√≥n por Edad --- \n")for edad, grupo in groupby(personas_ordenadas, key=key_edad):    print(f"Personas de {edad} a√±os:")    for persona in grupo:        print(f"  - {persona['nombre']} ({persona['ciudad']})")    print("---")
```
