# Unidad 01 - Strings en Python

## Crear un string de 4 caracteres y asignarlo a un nombre

```python
# Crear un string de 4 caracteres y asignarlo a un nombre
S = 'Spam'
print(S)
```

## Longitud del string

```python
len(S)
```

## Indexación - Primer elemento
En Python, los índices comienzan desde 0: el primer elemento está en el índice 0, el segundo en el índice 1, y así sucesivamente.

```python
S[0]
```

## Segundo elemento desde la izquierda

```python
S[1]
```

## Error al salirse de los límites

```python
# Si te sales de los límites
S[8]
```

## Indexación negativa - Último elemento
En Python, también podemos indexar hacia atrás desde el final.

```python
S[-1]
```

## Penúltimo elemento desde el final

```python
S[-2]
```

## Slicing (rebanado) de strings

```python
# desde la posición 1 hasta la 2 (no incluye la 3)
S[1:3]
```

```python
# Todo después del primer carácter (1:len(S))
S[1:]
```

```python
# Todo excepto el último carácter
S[:-1]
```

## Concatenación y repetición

```python
S + 'eggs'
```

```python
# Repetición
S * 6
```

## Inmutabilidad de los strings
Los strings en Python son inmutables, no se pueden modificar una vez creados.

```python
S[0] = 'z'  # Esto dará error
```

## Pero podemos crear nuevos strings

```python
S = 'z' + S[1:]
S
```

## Métodos de strings - Buscar

```python
# Encontrar la posición de una subcadena
S = 'SpamEggsSpam'
S.find('pa')
```

## Reemplazar

```python
S.replace('Eggs', 'Bacon')
```

## Mayúsculas y minúsculas

```python
S.upper()
```

```python
print("S es alfabético:", S.isalpha())
print("S es numérico:", S.isdigit())
```

## División de strings

```python
S = 'spams-spam-eggs-spam-bacon'
S.split('-')
```

## Formateo de strings

```python
'{}, eggs, and {}'.format('spam', 'SPAM!')
```

```python
S = 'SpamEggsSpam'
f'{S}'  # f-string (Python 3.6+)
```
